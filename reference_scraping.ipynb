```python
from bs4 import BeautifulSoup
import requests 
import pandas as pd
import requests
import json
import re
import textwrap

# This file manually retrieves the reference in a suitable format to be manually integrated
# This works for sagepub, semanticscholar, tandfondonline and springer links

# paper = link
paper = 'https://www.semanticscholar.org/paper/Environmental-Pressure-on-Software-as-a-Service-An-Kung-Cegielski/c9014b947cf39567ddc381962bfc0fb2d3b10afa'
response = requests.get(paper)
soup = BeautifulSoup(response.text, "html.parser")

if ('sagepub' in paper):
    print("sagepubscraping")
    table = soup.find("table", {"class": "references"})
    rows = table.find_all("tr")
    for j in range(0, len(rows)):
        tds = rows[j].find_all('td')
        reference = tds[1].text
        if (rows[j].find("span", {"class": "NLM_article-title"}) != None):
            title = rows[j].find("span", {"class": "NLM_article-title"}).text
            title_tag = '<t>' + title + '<t>'
            reference = reference.replace(title, title_tag)
        print((str(j+1) + ": " + reference).replace("\n", ""))

if ('semanticscholar' in paper):
    print("semanticscholarscraping")
    paper = paper.rsplit('/', 1)[-1]
    r = requests.get('https://api.semanticscholar.org/graph/v1/paper/' + paper + '?fields=title,references.title,references.authors,references.year,references.venue')
    r.json()
    data = r.text
    json_object = json.loads(data)
    references = json_object["references"]
    for j in range(0, len(references)):
        reference = ""
        for author in references[j]['authors']:
            reference += author['name'] + ", "
        reference += '<t>' + references[j]['title'] + '<t>, '
        reference += references[j]['venue'] + ", "
        reference += str(references[j]['year'])
        print (str(j+1) + ": " + reference)

if ('tandfonline' in paper):
    print("tandfonlinescraping")
    table = soup.find("ul", {"class": "references"})
    rows = table.find_all("li")
    for j in range(0, len(rows)):
        reference = rows[j].text
        if (rows[j].find("i") != None):
            journal = rows[j].find("i").text
            journal_tag = '<j>' + journal + '<j>'
            reference = reference.replace(journal, journal_tag)
        if (rows[j].find("span", {"class": "NLM_article-title"}) != None):
            title = rows[j].find("span", {"class": "NLM_article-title"}).text
            title_tag = '<t>' + title + '<t>'
            reference = reference.replace(title, title_tag)
        
        print(str(j+1) + ": " + reference)


if ("springer" in paper):
    print("springerscraping")
    table = soup.find("ol", {"class": "c-article-references"})
    if (table == None):
        table = soup.find("ol", {"class": "BibliographyWrapper"})
    rows = table.find_all("li")
    for j in range(0, len(rows)):
        if (rows[j].find("p", {"class": "c-article-references__text"}) != None):
            print(str(j+1) + ": " + rows[j].find("p", {"class": "c-article-references__text"}).text)
        else:
            print(str(j+1) + ": " + rows[j].find("div", {"class": "CitationContent"}).text)

    

```
