```python
#default journal adding
from csv import reader
from csv import DictReader
import re
import pandas as pd

additional_journals = []

filename = "./csv/additional_journals"
additional_journal_file = open(filename, "r")
for row in additional_journal_file:
    if row not in additional_journals:
        row = row.replace("\n", " ")
        if row != "":
            additional_journals.append(row)

columns = ['Name', 'Type']
df = pd.DataFrame(columns = columns)
for journal in additional_journals:
    values = [journal, "Journal"]
    df = df.append(pd.Series(values, index=columns), ignore_index=True)
df.to_csv('./csv/list_of_default_journals.csv', index=False, header=True)
```

```python
# classification files
unfiltered_paper = "./papers_curated/Wiener2020"
references = open(unfiltered_paper, "r")
classified_journal_lines = open("./classification_first_iteration/" + unfiltered_paper.split("/")[-1] + "_classified", 'w+')
unclassified_journal_lines = open("./classification_first_iteration/" + unfiltered_paper.split("/")[-1] + "_unclassified", 'w+')
used_journal_classifiers = open("./used_journal_classifiers/" + unfiltered_paper.split("/")[-1], 'w+')
journal_files = ["list_of_journals", "list_of_conferences" , "list_of_default_journals"]

# lists
scraping_keywords = ["sagepubscraping", "semanticscholarscraping", "tandfonlinescraping", "springerscraping"]
non_journals = ["harvard business review", "information systems", "big data"]
all_journals = []
journal_classifiers = []
```

```python
# journal aggregation

# functions
def has_brackets(journal):
    return bool(re.search(r"\((.*?)\)", journal))

def delete_brackets(journal):
    bracket_content = re.findall(r"\((.*?)\)", journal)
    for content in bracket_content:
        journal = journal.replace("(" + content + ")", "")
    return journal

def append_journal(journal, file):
    journal = journal.lower()
    if journal not in all_journals:
        if journal not in non_journals:
            if file != "list_of_default_journals":
                if len(journal.split()) > 1:
                    all_journals.append(journal)
            else:
                all_journals.append(journal)
                
# loop
for file in journal_files:
    with open("./csv/" + file + ".csv", "r") as read_obj:
        csv_dict_reader = DictReader(read_obj)
        for row in csv_dict_reader:
            journal = row['Name']
            append_journal(journal, file)
            if (has_brackets(journal)):
                journal = delete_brackets(journal)
                append_journal(journal, file)

                
print("Journal Count: " + str(len(all_journals)))
```

```python
# classification

# functions
def alphanumeric_filter (string):
    alphanumeric_filter = filter(str.isalnum, string)
    string = "".join(alphanumeric_filter)
    return string

def write_all_files (line, parent):
    if parent:
        classified_journal_lines.write("\n")
        unclassified_journal_lines.write("\n")
    classified_journal_lines.write(line)
    unclassified_journal_lines.write(line)
    
def delete_title_marker (line):
    return line.replace("<t>", "")

# loop
skip_until_next_parent = False
line_counter = 1
for line in references:
    print("line: " + str(line_counter))
    line_counter = line_counter + 1
    original_line = line
    
    #parent line
    if '<PARENTARTICLE>' in line:
        if '<NOPAPER>' in line:
            skip_until_next_parent = True
            continue
        else:
            skip_until_next_parent = False
            parent = True
            write_all_files(original_line, parent)
            continue

    #child of a non paper parent
    if skip_until_next_parent:
        continue
    
    #scraping_line
    if any(scraping_keyword in line for scraping_keyword in scraping_keywords):
        parent = False
        write_all_files(original_line, parent)
        continue

    #child line
    line = line.lower()
    line = alphanumeric_filter(line)
    
    match = False
    for journal in all_journals:
        if alphanumeric_filter(journal).lower() in line:
            match = journal 
            break
    if match:
        if match not in journal_classifiers:
            journal_classifiers.append(match)
        classified_journal_lines.write(original_line)
        continue

    original_line = delete_title_marker(original_line)
    unclassified_journal_lines.write(original_line)

classified_journal_lines.close()
unclassified_journal_lines.close()

for classifier in journal_classifiers:
    used_journal_classifiers.write(classifier)
    used_journal_classifiers.write("\n")
used_journal_classifiers.close()

print("finished")
```
